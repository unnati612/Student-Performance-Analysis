```python
import numpy as np
```

```python
np.random.seed(42)# starting point of random number generation
```

```python
students=100
```

```python
maths=np.random.randint(35,100,students)
science=np.random.randint(30,100,students)
english=np.random.randint(40,100,students)

marks=np.column_stack((maths,science,english))
marks[:5]

```

```python
mean_scores = np.mean(marks, axis=0)
median_scores = np.median(marks, axis=0)
std_scores = np.std(marks, axis=0)

subjects = ["maths", "science", "english"]

for i in range(3):
    print(f"{subjects[i]} → Mean: {mean_scores[i]:.2f}, "
          f"Median: {median_scores[i]}, Std Dev: {std_scores[i]:.2f}")
```

```python
total_scores = np.sum(marks, axis=1)

top_10 = np.percentile(total_scores, 90)
bottom_10 = np.percentile(total_scores, 10)

top_students = total_scores >= top_10
bottom_students = total_scores <= bottom_10

print("Top 10% cutoff:", top_10)
print("Bottom 10% cutoff:", bottom_10)
print(np.sum(top_students))
print(np.sum(bottom_students))
```

```python
pass_condition = np.all(marks >= 40, axis=1)

passed = np.sum(pass_condition)
failed = students - passed

print("Passed students:", passed)
print("Failed students:", failed)
```

```python
correlation_matrix = np.corrcoef(marks, rowvar=False)    #how strong 

print("Correlation Matrix:\n", correlation_matrix)
```

```python
import matplotlib.pyplot as plt

mean_total = np.mean(total_scores)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(total_scores, bins=10)

# Add vertical lines
plt.axvline(mean_total, linestyle='--', linewidth=2, color='red', label='Mean')
plt.axvline(top_10, linestyle=':', linewidth=2, color='black', label='10th Percentile')
plt.axvline(bottom_10, linestyle=':', linewidth=2, color='black', label='90th Percentile')

# Labels and title
plt.title("Histogram of Total Student Scores with Mean & Percentiles")
plt.xlabel("Total Score")
plt.ylabel("Number of Students")
plt.legend()
plt.grid(True)
plt.show()
```
